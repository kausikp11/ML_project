```python
import pandas as pd
import numpy as np
df = pd.read_csv('ISCX_Botnet-Training.csv')
```

```python
df2=df.drop(columns=['No.', 'Time'])
df2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df2['Protocol'])
label_protocol = le.transform(df2['Protocol'])
```

```python
le1 = LabelEncoder()
le1.fit(df2['Source'])
label_source = le1.transform(df2['Source'])

le2 = LabelEncoder()
le2.fit(df2['Destination'])
label_dest = le2.transform(df2['Destination'])

le3 = LabelEncoder()
le3.fit(df2['Info'])
label_info = le3.transform(df2['Info'])
print(label_dest,label_source,label_info)
```

```python
label_source = pd.Series(label_source.reshape(9388270))
label_dest = pd.Series(label_dest.reshape(9388270))
label_protocol = pd.Series(label_protocol.reshape(9388270))
label_info = pd.Series(label_info.reshape(9388270))
```

```python
df4 = pd.DataFrame([label_source,label_dest,label_protocol,label_info])
```

```python
df4 = df4.transpose()
df4.columns =['Source','Dest','Protocol','Info']
df4
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(df4)
```

```python
a = kmeans.labels_
```

```python
result = kmeans.predict(df4)
```
